```python
#Configuration
class CFG():
    seed = 42
    epochs = 8
    folds = 5
    batch_size = 6

    lr = 5e-6
    pct_start = 0.3
    max_lr = 1e-2
    weight_decay = 0.5
    shuffle = True
    
    def __init__(self, **kwargs):
        for k, v in kwargs.items():
            setattr(self, k, v)
            
#Setting a constant seed
def seed_everything(seed=42):
    imgaug.random.seed(seed)
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    
seed_everything(seed=42)
config = CFG()
```

```python
device = "cuda"
```

```python
#Create a Data class to hold the data and corresponding class labels
class GroceryStoreData(Dataset):
    #Create the constructor for the Data class
    def __init__(self, imgs, resize, width, height):
        #Data samples
        self.resize = resize
        self.width = width
        self.height = height
        
        self.img_paths = imgs

        
        
        #Classes for the labels (Background is first)
#         self.classes = ["_", "void"]
        self.classes = ["_", "void"]
        
        #Data Augmentation Parameters
        self.shift_x = np.linspace(0,80,10)
        self.shift_y = np.linspace(0,80,10)
        self.rot = np.linspace(0,30,10)
        self.shear = np.linspace(0,10,10)
        self.sola = np.linspace(0,256,10)
        self.post = [4,4,5,5,6,6,7,7,8,8]
        self.cont = [-0.2,0.2]
        self.bright = [-0.2, 0.2]
        self.shar = np.linspace(0.1,0.9,10)
        self.cut = np.linspace(0,60,10)

        self.N=2
        self.M=0
        self.p=0.4
        
        #FILES: MAKE SURE TO CHANGE THIS PATH
        self.files = glob.glob(f"{self.img_paths}/*.json")
        self.files = sorted(self.files, key = lambda x: int(x.split("/")[-1].split(".")[0].split("_")[-1]))
#         executor = ProcessPool(nodes = 8)
#         results = list(chain(executor.map(lambda f: self.processJson(f), files)))
        self.x = []
        self.y = []
        
        #Loading Data
        for f in self.files:
            self.processJson(f, num = 8)
        
        #Correcting for incorrect bounding box specification (PASCAL VOC)
        for i, box in enumerate(self.y):
            for bc in box['boxes']:
                if bc[0] > bc[2]:
                    bc[[0,2]] = bc[[2,0]]
                if bc[1] > bc[3]:
                    bc[[1,3]] = bc[[3,1]]


                    
                    
                # print(f"After {idx}!: {box}")

    #Return the length of the Data if len() is called on it  
    def __len__(self):
        return len(self.x)
                          
    # Decodes training images from the JSON and store the bounding box data for COCO evaluation metrics
    def processJson(self, file, num):
        f = open(file)
        data = json.load(f)
        
        img = io.BytesIO()
        img.write(base64.b64decode(data["imageData"]))
        x = np.array(PIL.Image.open(img))
        y = {}
        
        boxes = [list(np.sort(np.array(box["points"]), axis = 0).flatten()) for box in data["shapes"] if box["label"] == "void"]
        labels = [self.classes.index(box["label"]) for box in data["shapes"] if box["label"] == "void"]
#         y["labels"] = [box["label"] for box in data["shapes"] if box["label"] != "almost_empty"]
        
#         fig, ax = plt.subplots()
        
#         ax.imshow(x)
#         width = y["boxes"][0][2] - y["boxes"][0][0]
#         height = y["boxes"][0][3] - y["boxes"][0][1]
#         rect = patches.Rectangle((y["boxes"][0][0], y["boxes"][0][1]), width, height, linewidth=2, edgecolor='r')
#         ax.add_patch(rect)
#         plt.show()

        teacher_transform = A.Compose(
            [
                A.augmentations.geometric.resize.LongestMaxSize(max_size = self.resize),
                A.RandomCrop(width=self.width, height=self.height, p=1),
                A.ShiftScaleRotate(shift_limit_x=self.shift_x[self.M], rotate_limit=15, shift_limit_y=self.shift_y[self.M], p=self.p),

                A.Equalize(p=self.p),
                A.RandomBrightnessContrast(brightness_limit = self.bright, contrast_limit = self.cont, p = self.p),
            ], bbox_params=A.BboxParams(format='pascal_voc', label_fields = ["class_labels"], min_visibility = 0.7)
        )
        
        for i in range(num):
            transformed = teacher_transform(image=x, bboxes=boxes, class_labels = labels)
            aug_x = torch.reshape(torch.FloatTensor(transformed['image']), (3, 512, 512))

            bboxes = torch.reshape(torch.FloatTensor(transformed['bboxes']), (-1, 4))
            classes = torch.reshape(torch.LongTensor(transformed['class_labels']), (-1, ))
            aug_y = {
                'boxes': bboxes,
                'labels': classes,
                'image_id': torch.tensor([self.files.index(file) * 8 + i]),
                'area': torch.abs(bboxes[:, 3] - bboxes[:, 1]) * torch.abs(bboxes[:, 2] - bboxes[:, 0]),
                'iscrowd': torch.zeros(len(bboxes))
            }
            
            self.x.append(aug_x)
            self.y.append(aug_y)
        
                        
    # Correcting for incorrect bounding boxes                    
    def __getitem__(self, idx):
        for box in self.y[idx]["boxes"]:
            if box[0] > box[2]:
                box[[0,2]] = box[[2,0]]
            if box[1] > box[3]:
                box[[1,3]] = box[[3,1]]
        return self.x[idx], self.y[idx]
```

```python
def custom_collator(batch):
    x, y = [], []
    
    for (image, labels) in batch:
        x.append(image / 255.0)
        y.append(labels)

    return x, y
```

```python
# Correcting for incorrect bounding boxes      
def fix_data(data_loader):
    for x,y in data_loader:
        for data in y:
            for bc in data["boxes"]:
                if bc[0] > bc[2]:
                    bc[[0,2]] = bc[[2,0]]
                if bc[1] > bc[3]:
                    bc[[1,3]] = bc[[3,1]]
```

```python
def train_loop(data, config):
    # Start K-Fold Cross Validation
    kf = KFold(n_splits = config.folds, shuffle = True, random_state = config.seed)
    total_metrics = []
    
    for fold, (train_idx, val_idx) in enumerate(kf.split(np.arange(len(data)))):
        print(f"*********************** Fold: {fold} ***********************")
        gc.collect()
        # Using Swin-T Transformer as a backbone to extract context
        backbone = torchvision.models.swin_t(weights = torchvision.models.Swin_T_Weights.DEFAULT).features
        
        backbone.out_channels = 25
        # Due to smaller voids at times, we chose to start with 4 as a bounding box size
        anchor_generator = torchvision.models.detection.rpn.AnchorGenerator(sizes=((4, 8, 16, 32, 64, 128),),
                                   aspect_ratios=((0.5, 1.0, 2.0),))
        roi_pooler = torchvision.ops.MultiScaleRoIAlign(featmap_names=['0'],
                                                output_size=7,
                                                sampling_ratio=2)
        
        # Faster RCNN head
        model = torchvision.models.detection.FasterRCNN(backbone,
                   num_classes=2,
                   rpn_anchor_generator=anchor_generator,
                   box_roi_pool=roi_pooler).to(device)

        optimizer = torch.optim.AdamW(model.parameters(), lr=config.lr, weight_decay = config.weight_decay)

        # 1CycleLearning Scheduler
        scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr = config.max_lr, epochs = config.epochs, steps_per_epoch = len(data) // config.batch_size, pct_start = config.pct_start)
        
        epoch_metrics = []
        train_sampler = SubsetRandomSampler(train_idx)
        val_sampler = SubsetRandomSampler(val_idx)
        
        train_loader = DataLoader(data, batch_size = config.batch_size, sampler=train_sampler, collate_fn = custom_collator)
                
        val_loader = DataLoader(data, batch_size = config.batch_size, sampler=val_sampler, collate_fn = custom_collator)

#         train_loader = DataLoader(data, batch_size = config.batch_size, collate_fn = custom_collator)
        for epoch in tqdm(range(config.epochs)):
            print(colored(f'Starting Epoch: {epoch}', 'red'))
            print()
            model.train()

            train_one_epoch(model, optimizer, train_loader, device, epoch, print_freq=10)
            
            scheduler.step()
            
            torch.cuda.empty_cache()
            
            print()
            print(colored(f'Evaluating Epoch: {epoch}', 'green'))
            model.eval()
            
            metrics = evaluate(model, val_loader, device=device)
            fix_data(train_loader)
            fix_data(val_loader)
            print("\n\n\n")
            epoch_metrics.append(metrics)
            
        torch.save(model, f"model_fold_{fold}.pth")
        total_metrics.append(epoch_metrics[fold])
        del model
        del optimizer
        del backbone
        del anchor_generator
        del roi_pooler
        del scheduler
    return total_metrics
    
```

```python
##### import gc
gc.collect()
metrics = train_loop(data, config)
```

```python
def nms(y_hat, iou_thresh=0.5):
    
    keep = torchvision.ops.nms(y_hat['boxes'], y_hat['scores'], iou_thresh)
    
    final_prediction = y_hat
    final_prediction['boxes'] = final_prediction['boxes'][keep]
    final_prediction['scores'] = final_prediction['scores'][keep]
    final_prediction['labels'] = final_prediction['labels'][keep]
    
    return final_prediction
```
