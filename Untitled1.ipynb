```python
to_delete = []
import pdb
for i,box in enumerate(boxes):
    # fix box coordinate order

    for bc in box['boxes']:
        if bc[0] > bc[2]:
            bc[[0,2]] = bc[[2,0]]
        if bc[1] > bc[3]:
            bc[[1,3]] = bc[[3,1]]
        if bc[0] == bc[2] or bc[1] == bc[3]:
            to_delete.append(i)
            
print(boxes)
        
```
